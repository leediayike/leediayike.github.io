```python
import altair as alt
import pandas as pd

df = pd.read_csv("https://github.com/UIUC-iSchool-DataViz/is445_data/raw/main/licenses_fall2022.csv", parse_dates=['Original Issue Date', 'Effective Date'])

chart1 = alt.Chart(df).mark_circle(size=60).encode(
    x='year(Original Issue Date):O',
    y='License Type:N',
    color='License Status:N',
    tooltip=['License Type', 'License Status', 'Original Issue Date']
).properties(
    width=500,
    height=300,
    title='License Type vs. Issue Year by Status'
).interactive()

license_dropdown = alt.binding_select(options=df['License Type'].unique())
license_select = alt.selection_point(fields=['License Type'], bind=license_dropdown, name='SelectType')

interactive_chart = chart1.add_params(license_select).transform_filter(license_select)
```

```python
alt.data_transformers.disable_max_rows()
```

```python
interactive_chart.save('../assets/json/licenses.json')
```

```python
hist = alt.Chart(df).mark_bar().encode(
    x=alt.X('year(Original Issue Date):O', title='Year'),
    y='count()',
    color='License Status:N',
    tooltip=['License Status', 'count()']
).properties(
    width=500,
    height=300,
    title='Number of Licenses Issued per Year by Status'
)

hist.save('../assets/json/licenses_hist.json')
```
